```python
# Importing necessary libraries
from bs4 import BeautifulSoup
import requests
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
import numpy as np
import pandas as pd
import re
import time
```

```python
URL = "https://www.tripadvisor.com/Restaurants-g294217-Hong_Kong.html"
driver = webdriver.Chrome(executable_path=r"C:\Users\Admin\Desktop\chromedriver")
driver.get(URL)
restaurants = pd.DataFrame()
time.sleep(5)
subhtml = driver.page_source
soup = BeautifulSoup(subhtml, "html.parser")

#while restaurants.shape[0]<400:
while True:
    for i in soup.find_all(class_="_1llCuDZj"):
        try:
            rank_name = i.find(class_="_15_ydu6b").text
            try:
                rank = int(re.findall("(\d*)\.", rank_name)[0])
            except:
                rank = np.nan
            try:
                name = re.findall("\.\s(.*)", rank_name)[0]
            except:
                name = np.nan
        except:
            rank = np.nan
            name = np.nan
        try:
            rating = i.find("svg", class_="_3KcXyP0F")["title"]
            rating = float(re.findall("(.*) of 5 bubbles", rating)[0])
        except:
            rating = np.nan
        try:
            total_review = i.find(class_="w726Ki5B").text.split()[0]
        except:
            total_review = np.nan
        try:
            genre = np.nan
            price = np.nan
            for j in i.find(class_="MIajtJFg _1cBs8huC _3d9EnJpt").find_all(class_="_1p0FLy4t"):
                if '$' in j.text:
                    price = j.text
                else:
                    genre = j.text
        except:
            genre = np.nan
            price = np.nan
        try:
            safety = i.find(class_="_1shPZD63").text
        except:
            safety = np.nan
        try:
            review_1 = i.find_all(class_="_1OfeygW_")[0].text
        except:
            review_1 = np.nan
        try:
            review_2 = i.find_all(class_="_1OfeygW_")[1].text
        except:
            review_2 = np.nan
        restaurants = restaurants.append({"review_2": review_2, "review_1": review_1, "safety": safety, 
                                         "genre": genre, "price": price, "total_review": total_review, "rating": rating, 
                                          "name":name, "rank": rank}, ignore_index = True)
    
    nextlink = soup.find(class_='nav next rndBtn ui_button primary taLnk')
    if nextlink:
        nextlink = URL + nextlink['href']
        subhtml = driver.get(nextlink)
        print(nextlink)
        
        time.sleep(5)
        subhtml = driver.page_source
        soup = BeautifulSoup(subhtml, 'html.parser')
        
    else:
        break


restaurants
        
        
    
        
    
        
        
        
            
            
            
```
